Vamos considerar um Dataframe simples com os países onde o Titanic fez paragens.

```python
cities = ['Cherbourg', 'Queenstown', 'Southampton']
countries = ['France', 'Ireland', 'United Kingdom']
gnp_1925 = [36262, 1862, 43700] # produto nacional bruto em 1925 medido em milhões de USD de 1960
life_expectancy_1926 = [54, 59.6, 54] 

dict = {'city': cities, 'country': countries, 'gnp': gnp_1925, 'life_expectancy':life_expectancy_1926} 
    
countries_df = pd.DataFrame(dict)
countries_df
```

Vamos juntar os dois dataframes, e calcular mais algumas métricas.

**P9: Faça a junção de `titanic_df` com `countries_df` sobre a coluna `embark_town`.**

```python

combined_df = titanic_df.merge(countries_df, left_on="embark_town", right_on="city", how="outer")
```

```python
combined_df
```

**P10: Adicione uma coluna `life_stage`, em `combined_df` com a percentagem de vida vivida de cada passageiro, atendendo à sua idade e esperança de vida. Assuma que a nacionalidade do passageiro é a que corresponde ao país onde embarcou**

```python
combined_df['life_stage'] = (combined_df["age"] / combined_df["life_expectancy"]) * 100 
```

```python
combined_df
```

**P11: Crie um histograma que permita visualizar de forma agregada a idade dos passageiros, em 8 grupos. Utilize o estilo "darkgrid" do seaborn, e dê ao gráfico o título "Distribuição das Idades dos Passageiros".**

```python
sns.set_style("darkgrid")

plt.hist(titanic_df["age"], bins=8, color="red")
plt.title("Distribuição das Idades dos Passageiros", fontsize=16)
plt.show()
```
